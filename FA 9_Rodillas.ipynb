```python
#9. Analyze songs by artist and views-youtube, stream-spotify.
df.groupby('Artist')[['Views', 'Stream']].sum()
```

```python
#10. Create bar charts for Top 10 Artists on Youtube and Top 10 Artists on Spotify in 1 row of display.
#11. Make some remarks about the Top 10 Artists on Youtube and Top 10 Artists on Spotify.

#YT views top 10
df.groupby('Artist')[['Views']].sum().sort_values(['Views']).head(10)

#Spotify streams top 10
df.groupby('Artist')[['Stream']].sum().sort_values(['Stream']).head(10)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))

#YT graph
ax1.set_title('Top 10 Artists on YouTube')
df.groupby('Artist')[['Views']].sum().sort_values(['Views'], ascending=False).head(10).plot(kind='bar', ax=ax1)

#Spotify graph
ax2.set_title('Top 10 Artists on Spotify')
df.groupby('Artist')[['Stream']].sum().sort_values(['Stream'], ascending=False).head(10).plot(kind='bar', ax=ax2)


ax1.set_ylabel('Views per ten billions')
ax2.set_ylabel('Streams per ten billions')
plt.show()
```

The figures above shows the Top 10 Artists for two platforms, both the Youtube views and Spotify streams, both per 10 billions.

One the left figure, Ed Sheeran holds the top spot with an accumulated number of views at a little over 15 billion views. Followed by CoComelon, a children-based youtube page with views over 14 billion views. The two Youtube giants are then followed by Katy Perry and Charlie Puth who both broke 12 billion views at 13 billion and 12.2 billion views respectively. Luis Fonsi rounds out the top 5 artist who are just below the 12 billion views mark at around 1.1 billion views. The top 10 Youtube music artist is completed by Justin Bieber, Daddy Yankee, Bruno Mars, Macklemore & Ryan Lewis, and Coldplay, who are all above the 10 billion mark. 

On the other hand, the right figure shows artist with the most Spotify streams. It is dominated by the American rapper and producer Post Malone with around 15 billion streams, followed by the #1 Youtube artist Ed Sheeran, with 14 billion streams on his Spotify. The two are followed by Dua Lipa, XXXTENTACION, and The Weeknd who were all above 13 billion streams on Spotify, rounding out the top 5 Artists on the platform. It is then followed by Justin Bieber at 6, who were also 6th in Youtube streams, Imagine Dragons, Coldplay, Khalid, and Bruno Mars, all hovers around 10 billion to 12 billion streams. 

The two graphs represents the difference of popularity of different artists in two different platforms.

```python
#16. Determine the top 10 artists on the Youtube platform whose songs have more likes and display a bar chart with a title Number of Likes per Artist and make some remarks.

df_artist = df.set_index("Artist")
sns.barplot(x="Likes", y="Artist", data=df_artist.groupby("Artist")["Likes"].sum().reset_index().sort_values(by="Likes", ascending=False).head(10))
plt.xlabel("Number of Likes")
plt.ylabel("Artist")
plt.title("Number of Likes per Artist")
plt.show()
```

On a close second, Korean's BLACKPINK is the second most liked artist who garnered over 120 million likes on their official Youtube page. The two Korean titans dominated the most liked aspect on the Youtube platform, which is followed by Charlie Puth, Ed Sheeran, and Justin Bieber, all who has likes that ranges from 70 to 80 million likes, completing the top 5 most liked artists. These 5 are then followed by artists Billie Eilish, Imagine Dragons, Alan Walker, DJ Snake, and Luis Fonsi, who all are over 60 million likes. 

Based on the graph above, it is safe to assume that the two Korean musical powerhouses, BTS and BLACKPINK, are not only popular in their country, but is also a well-established name all around the globe, and one of the most popular in our century.

```python
#20. Identify the top 10 channels based on views and display a bar chart of it and make some remarks.

sns.barplot(x="Views", y="Channel", data=df.groupby("Channel")["Views"].sum().reset_index().sort_values("Views", ascending=False).head(10))
plt.title('Top Youtube Channels based on Views')
plt.xlabel('Views')
plt.ylabel('Channel')
plt.show()
```

The table given above is a representation of the top Youtube channels currently on the platform based on views. In a dominating fashion, the Indian-based platform T-Series almost doubles every other channels on the graph by accumulating around 50 billion views on the platform. It is followed by the VEVO channels of LuisFonsi and Calvin Harris with both of it garnering around 20 million views. The top 5 Youtube channels is completed by Macklemore and Ed Sheeran with almost an equal amount of views at around 15 billion. Subsequently, Hear This Music, Cocomelon channel, Bruno Mars, KatyPerryVEVO, and DJSnakeVEVO completes the top youtube channels on the platform based on its accumulated views. 

Although T-Series dominated this graph, it is a result of a prominent posting of many artists and their videos, different from other channels who mainly sourced from a single artist.
