```python
def missing(x):
    missing_values_count = x.isnull().sum().sum()
    return missing_values_count

def Coefficient_of_Variation(x):
    return x.std() / x.mean()
def Std_Error_Skewness(x):
    SES = (6*x.count()*(x.count()-1) / ((x.count()-2)*(x.count()+1)*(x.count()+3)))**0.5
    return SES
def Std_Error_Kurtosis(x):
    SES = (6*x.count()*(x.count()-1) / ((x.count()-2)*(x.count()+1)*(x.count()+3)))**0.5
    return ( 4*(x.count()**2-1) * ((6*x.count()*(x.count()-1) / ((x.count()-2)*(x.count()+1)*(x.count()+3)))**0.5)**2 / ((x.count()-3)*(x.count()+5)) )**0.5


Percentiles = [25,50,75,10,20,30,40,50,60,70,80,90]
def Percentiles_25th(x):
    return np.percentile(x, 25)
def Percentiles_50th(x):
    return np.percentile(x, 50)
def Percentiles_75th(x):
    return np.percentile(x, 75)
def Percentiles_10th(x):
    return np.percentile(x, 10)
def Percentiles_20th(x):
    return np.percentile(x, 20)
def Percentiles_30th(x):
    return np.percentile(x, 30)
def Percentiles_40th(x):
    return np.percentile(x, 40)
def Percentiles_50th(x):
    return np.percentile(x, 50)
def Percentiles_60th(x):
    return np.percentile(x, 60)
def Percentiles_70th(x):
    return np.percentile(x, 70)
def Percentiles_80th(x):
    return np.percentile(x, 80)
def Percentiles_90th(x):
    return np.percentile(x, 90)
```
